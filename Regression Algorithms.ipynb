```python
import pandas as pd #importing pandas 
```

```python
print(pd.__version__) #the version of pandas we are using is 0.25.0
```

```python
#we read the data file with no specified headers and we assign name to the columns that do not possess header originally.
auto_data = pd.read_csv('imports-85.data', header = None, names = ['symboling', 'normalized_losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine_type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price'])
```

```python
auto_data #we see, how the data looks like. 
```

```python
#As we can see, there are lots of ? in the data, we replace those values by NaN value from Numpy. 
import numpy as np
auto_data = auto_data.replace('?', np.nan)
auto_data.head() #we see the first five rows to see, if the changes are made. 
```

```python
#We see how the overall data is distributed
auto_data.describe(include = 'all').T #include = 'all' shows description of all the columns
```

```python
auto_data['price'].describe() #We check the datatype of the price column. 
```

```python
#since the data-type is object, we need to convert it to numeric in order to use it for prediction. 
auto_data['price'] = pd.to_numeric(auto_data['price'], errors = 'coerce') # The errors = 'coerce' neglects the errors. 
```

```python
#Now, we can see the price column is converted to numeric values and see some information in that. 
auto_data['price'].describe()
```

```python
auto_data = auto_data.drop('normalized_losses', axis = 1) #We drop the normalized-losses column as it doesn't have significant contribution to price. 
```

```python
auto_data.head() #We check if the normalized_losses column is well removed. 
```

```python
auto_data.columns #We list all the columns in the dataframe. 
```

```python
auto_data['horsepower'].describe() #We see the details of horsepower column. 
```

```python
#As the data-type of the horsepower column is object, we also convert it to numeric type. 
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors = 'coerce')
```

```python
#We similarly check a categorical value in column num-of-cylinders. 
auto_data['num-of-cylinders'].describe()
```

```python
#We check the unique values in the num-of-cylinders column. 
auto_data['num-of-cylinders'].unique()
```

```python
#Since, the num-of-cylinders column is in string, we change them to numeric form by creating a dictionary value and replacing the column values in the dataframe. 
num_dict = {'four':4, 
           'six': 6,
           'five':5,
           'three':3,
           'twelve':12,
           'two':2,
           'eight':8}
auto_data['num-of-cylinders'].replace(num_dict, inplace = True) #we replace the column values with the dictionary we created. 
```

```python
#we check if the conversion is reflected correctly. 
auto_data['num-of-cylinders'].describe()
```

```python
#Now, we change the other categorical values to numeric values with pandas get_dummies function which creates a numeric value of columns to use it for prediction. 
auto_data = pd.get_dummies(auto_data, 
                          columns = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine_type', 'fuel-system'])
```

```python
#we look few rows of the dataframe
auto_data.head()
```

We can see that dummy columns are inserted with numeric values. Now we can use the machine learning algorithms for prediction purpose. 

```python
#We remove the rows having null values in the dataframe. 
auto_data = auto_data.dropna()
auto_data.shape
```

```python
#we check if there are any rows still remaining with the null values. 
auto_data[auto_data.isnull().any(axis = 1)]
```

```python
#we import the train_test_split module from sklearn which we'll use to divide our dataframe to training and testing set. 
from sklearn.model_selection import train_test_split
X = auto_data.drop('price', axis = 1) #since we're using price as a value to be predicted, we remove that column and assign remaining values to training dataset.
y = auto_data['price'] #the predicted value for the dataframe.
#we divide the dataset into training and test dataset, the test-size 0.2 states 20% of the data will be used for testing while the remaining 80% will be used for training. 
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#we first of all import LinearRegression module from sklearn. 
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, Y_train) #we fit the training data into the LinearRegression. 
```

```python
#we calculate the score, which gives the coefficient of R^2 score, which is the variance of fitted values divided by the variance of actual values. 
reg.score(X_train, Y_train)
```

```python
predictors = X_train.columns #we assign the columns of the training dataset to predictors. 
pd.set_option('display.max_rows', 500) #to display all the possible values. the number 500 states the no. of rows shown will be 500. 
coef1 = pd.Series(reg.coef_, predictors).sort_values() #we use the coefficient of our linear regression and we match it with the columns and view it in descending order. 
print(coef1)
```

```python
#We calculate the predicted value of the testing part of the dataframe. 
y_pred = reg.predict(X_test)
```

```python
#we plot our predicted value alongside the actual value. 
import matplotlib.pyplot as plt
plt.plot(y_pred, label = 'predicted')
plt.plot(Y_test.values, label = 'actual')
plt.legend()
plt.show()
```

```python
#we calculate the R^2 score again. 
r_square = reg.score(X_test, Y_test)
r_square
```

We can see that the score is 0.63 which is not really good. We'll see the score with other algorithms too. The maximum value of the score is 1.0

```python
#we see the total mean squared error of our algorithm.  
from sklearn.metrics import mean_squared_error
linear_model_mse = mean_squared_error(y_pred, Y_test) #we feed the predicted result and calculate the error with the actual values. 
linear_model_mse
```

```python
#we calculate the root mean squared value which is the square root of mean squared value. 
import math
rmse = math.sqrt(linear_model_mse)
rmse
```

The root mean square value shows by how much the predicted price are deviated from the actual price in average. 

```python
#we replicate the above problem with Lasso regression. 
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha = 0.5, normalize = True) #For lasso, we need to specify the hyperparameter. we specify the value of alpha to be 0.5 and now we normalize our model. 
lasso_model.fit(X_train, Y_train) #we fit the training data into the lasso model. 
```

```python
#we calculate the score of the lasso model. 
lasso_model.score(X_train, Y_train)
```

We can see that the score is increased and the model is performing very well than Linear Regression due to the penalizing(which is covered in theoretical part of Lasso regression)

```python
#we see the significance of columns in our model. The lower value represents lower significance and higher value represents higher significance. 
coef = pd.Series(lasso_model.coef_, predictors)
print(coef)
```

```python
#we calculate the predicted values in the testing dataset. 
y_predict = lasso_model.predict(X_test)
```

```python
#we plot the predicted values with the actual values. We should see that our regression line is well fitted than in the earlier case. 
plt.plot(Y_test.values, label = 'actual_values')
plt.plot(y_predict, label = 'predicted values')
plt.legend()
plt.show()
```

```python
#We calculate the R^2 score of the lasso regression. 
r2_lasso = lasso_model.score(X_test, Y_test)
r2_lasso
```

```python
#We calculate the mean squared error. 
lasso_mse = mean_squared_error(y_predict, Y_test)
lasso_mse
```

```python
#we calculate the root mean squared error.
rmse_lasso = math.sqrt(lasso_mse)
rmse_lasso
```

We can see that the root mean squared error is well reduced than the earlier Linear Regression. We can conclude this regression is better fit than the previous regression in this particular problem.

```python
#We replicate the same problem with Ridge regression. 
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha = 0.5, normalize = True) #we use the same coefficients as we used in Lasso Regression. 
ridge_model.fit(X_train, Y_train) #we fit the training data into our model. 
```

```python
#we calculate the R^2 score on our training dataset. 
ridge_model.score(X_train, Y_train) 
```

```python
#we see the weight of the features we have in our dataframe. 
coef_ridge = pd.Series(ridge_model.coef_, predictors).sort_values()
coef_ridge
```

```python
#we calculate the predicted values or y_values of our training dataset. 
y_ridge = ridge_model.predict(X_test)
```

```python
#we plot the predicted values and the actual values. we can see the data is well fitted. 
plt.plot(Y_test.values, label = 'actual values')
plt.plot(y_ridge, label = 'predicted values')
plt.legend()
plt.show()
```

```python
#we calculate the R^2 score on our test dataset. We can say this model is well fitted than the earlier two models. 
ridge_model.score(X_test, Y_test)
```

```python
#We calculate the mean squared error. 
mse_ridge = mean_squared_error(y_ridge, Y_test)
mse_ridge
```

```python
#we calculate the root mean squared error. 
rmse_ridge = math.sqrt(mse_ridge)
rmse_ridge
```

We can see that the root mean squared error value is pretty less compared to the other two, which shows that our model is well fitted for training as well as for test dataset. 

```python
#we finally use support vector machine regression approach. 
from sklearn.svm import SVR
svr_model = SVR(kernel = 'linear', C = 1.0) #the kernel linear states it is the linear regression and the C value is for penalizing the regression line. 
svr_model.fit(X_train, Y_train) #we fit the training dataset into our model. 
```

```python
svr_model.score(X_train, Y_train) #we calculate the R^2 score of the training dataset. we can see it is lower than the earlier datasets. 
```

```python
#we calculate the significance of the columns in the dataframe. 
coef_svr = pd.Series(svr_model.coef_[0], predictors).sort_values()
coef_svr
```

```python
#we predict the y_values of our testing dataset using svm module. 
y_svr = svr_model.predict(X_test)
```

```python
#we plot the predicted values and the actual values. The fit seems okay, but it's not better as previous algorithms. 
plt.plot(Y_test.values, label = 'actual')
plt.plot(y_svr, label = 'predicted')
plt.legend()
plt.show()
```

```python
#We calculate the score of R^2 in the test dataset. 
svr_model.score(X_test, Y_test)
```

```python
#we calculate the mean squared error. 
mse_svr = mean_squared_error(y_svr, Y_test)
mse_svr
```

```python
#we calculate the root mean squared error. 
rmse_svr = math.sqrt(mse_svr)
rmse_svr
```

We can see that the root mean squared value is not very well, which means our predicted values is deviated by this amount than the actual values. 

From this approach, we found out Ridge regression to be the best fitting model for our problem. However, hyperparameter plays a vital role in selecting those algorithms and optimizing the hyperparameter is also a crucial stage in using these algorithms. This notebook, however, gives an intuition on algorithms that can be used for regression problems.
